# RMS Prop

**Introduction**

En deep learning, les optimisateurs jouent un rôle crucial dans l'entraînement des modèles de réseau de neurones. Leur objectif est d'ajuster les poids et les biais du modèle afin de minimiser la fonction de perte et d'améliorer ses performances. Parmi les nombreux optimisateurs disponibles, RMSprop est largement utilisé et apprécié pour ses capacités d'adaptation des taux d'apprentissage.

Dans cette présentation, nous allons examiner de plus près le fonctionnement de RMSprop, ses avantages et ses limites, et le comparer à d'autres optimisateurs populaires. Nous illustrerons également son efficacité en fournissant un exemple concret où RMSprop permet d'obtenir de meilleurs résultats en termes de précision (accuracy) lors de l'entraînement d'un modèle de réseau de neurones.

**Explication de RMSprop**

RMSprop, ou Root Mean Square Propagation, est un optimisateur utilisé en deep learning qui se distingue par sa capacité à ajuster de manière adaptative les taux d'apprentissage pour chaque paramètre du modèle. Son fonctionnement repose sur deux principaux mécanismes : l'adaptation des taux d'apprentissage et l'utilisation d'une moyenne mobile exponentielle des carrés des gradients.

Pour l'adaptation des taux d'apprentissage, RMSprop utilise une approche qui permet de prendre en compte les variations des gradients entre les différents paramètres du modèle. Contrairement à d'autres optimisateurs qui utilisent un taux d'apprentissage global fixe, RMSprop ajuste le taux d'apprentissage individuel pour chaque paramètre en se basant sur les gradients précédents. Ainsi, les paramètres avec des gradients plus importants auront des taux d'apprentissage plus petits, tandis que les paramètres avec des gradients plus faibles auront des taux d'apprentissage plus élevés.

En ce qui concerne la normalisation des gradients, RMSprop utilise une moyenne mobile exponentielle des carrés des gradients précédents. Cette moyenne mobile est mise à jour à chaque étape d'apprentissage en utilisant un facteur d'amortissement (généralement proche de 0,9). En divisant le gradient actuel par la racine carrée de cette moyenne mobile, RMSprop parvient à normaliser les gradients et à atténuer les variations importantes. Cela permet d'éviter que les mises à jour des paramètres ne soient trop influencées par des gradients bruyants ou des variations importantes des gradients.

La formule de RMSprop pour la mise à jour des poids d'un modèle de réseau de neurones est la suivante:

g(t) = α * g(t-1) + (1 - α) * (∇J)^2

θ(t+1) = θ(t) - (η / √(g(t) + ε)) * ∇J

où :

* g(t) représente la moyenne mobile exponentielle des carrés des gradients précédents à l'instant t.
* α est le facteur d'amortissement qui contrôle le taux de décroissance de la moyenne mobile exponentielle (typiquement proche de 0.9).
* ∇J est le gradient de la fonction de perte par rapport aux poids θ.
* η est le taux d'apprentissage.
* ε est une petite valeur (par exemple, 10^-8) ajoutée au dénominateur pour éviter la division par zéro.

**Exemple d'application** :

Supposons que nous entraînons un modèle de réseau de neurones pour la classification d'images en utilisant RMSprop comme optimisateur. À chaque étape d'entraînement, nous calculons le gradient (∇J) par rapport aux poids du modèle. Voici un exemple fictif simplifié avec un seul poids θ et un pas d'apprentissage η de 0.01 :

À l'étape initiale (t = 0), supposons que g(t-1) = 0.

Lors de la première itération (t = 1), supposons que le gradient (∇J) calculé est -0.4.

En utilisant la formule de RMSprop, nous mettons à jour la moyenne mobile exponentielle des carrés des gradients :

g(t) = 0.9 * 0 + 0.1 * (-0.4)^2 = 0.016.

Ensuite, nous mettons à jour le poids θ en utilisant la formule :

θ(t+1) = θ(t) - (0.01 / √(0.016 + 10^-8)) * (-0.4) = θ(t) + 0.0079.

Ce processus se répète à chaque itération de l'entraînement, où les gradients sont mis à jour en utilisant RMSprop et les poids sont ajustés en conséquence. Cette approche adaptative des taux d'apprentissage et la normalisation des gradients permettent à RMSprop de mieux gérer les variations importantes des gradients et d'améliorer la convergence du modèle.

**Avantages de RMSprop**

RMSprop présente plusieurs avantages par rapport à d'autres optimisateurs, notamment en termes de robustesse face aux variations importantes des gradients et de gestion des surfaces de perte non convexes. De plus, RMSprop est souvent moins sensible aux choix des hyperparamètres, ce qui simplifie le processus d'optimisation.

* La robustesse: l'un des avantages clés de RMSprop est sa capacité à gérer efficacement les variations importantes des gradients entre les différents paramètres du modèle. En adaptant de manière adaptative les taux d'apprentissage pour chaque paramètre, RMSprop permet d'ajuster les mises à jour des poids en fonction des caractéristiques locales de la fonction de perte. Cela lui confère une plus grande robustesse face aux gradients bruyants ou aux variations significatives des gradients.


* La capacité de gérer les pertes: RMSprop se révèle particulièrement efficace lors de l'optimisation de surfaces de perte non convexes. Les surfaces de perte non convexes peuvent présenter de nombreux minimums locaux, ce qui rend l'optimisation plus complexe. RMSprop, avec sa capacité à s'adapter aux variations locales des gradients, peut mieux naviguer dans ces paysages de perte complexes et trouver des solutions de meilleure qualité.


* La sensibilité aux choix des hyperparamétres: en ce qui concerne les hyperparamètres, RMSprop est souvent moins sensible à leur choix que certains autres optimisateurs. Par exemple, le choix du taux d'apprentissage initial peut être moins critique dans RMSprop car il est adapté de manière adaptative à chaque paramètre. Cela réduit la dépendance à des réglages précis des hyperparamètres et permet une convergence plus efficace du modèle.




**Inconvénients de RMSprop**

Il existe également quelques inconvénients associés à cet optimisateur en deep learning :

* Sensibilité aux choix des hyperparamètres : Bien que RMSprop soit généralement moins sensible aux hyperparamètres que certains autres optimisateurs, il peut encore nécessiter un réglage soigneux des paramètres tels que le facteur d'amortissement α et la petite valeur ε ajoutée au dénominateur. Des choix inappropriés de ces hyperparamètres peuvent entraîner une convergence plus lente ou une performance inférieure du modèle.

* Accumulation d'informations historiques : RMSprop utilise une moyenne mobile exponentielle pour calculer la moyenne des carrés des gradients précédents. Cela signifie que les informations historiques sont prises en compte dans les mises à jour des paramètres. Cependant, dans certaines situations, l'accumulation excessive d'informations passées peut ralentir la convergence et entraîner une performance suboptimale.

* Manque de convergence garantie : Bien que RMSprop améliore généralement la convergence des modèles en deep learning, il ne garantit pas la convergence vers un minimum global. Comme pour tout optimisateur, RMSprop peut être sensible à l'initialisation des poids et aux propriétés spécifiques du problème. Il est donc important d'effectuer plusieurs essais avec différentes initialisations pour obtenir des résultats optimaux.

* Dépendance à la taille des mini-lots : RMSprop peut être sensible à la taille des mini-lots utilisés lors de l'entraînement. Des mini-lots trop petits peuvent entraîner une estimation peu fiable de la moyenne mobile des carrés des gradients, tandis que des mini-lots trop grands peuvent entraîner une adaptation lente des taux d'apprentissage. Le choix optimal de la taille des mini-lots peut varier en fonction du problème spécifique et peut nécessiter des ajustements expérimentaux.



**Application de RMSprop avec les données du diabéte**

```python
import numpy as np
import pandas as pd
import tensorflow as tf

df = pd.read_csv('diabetes.csv');
df.head()
```

```python
Y = df['Outcome']
X = df.drop('Outcome', axis=1)
```

```python
X.shape
```

```python
Y
```

```python
# Normalisation des caractéristiques
from sklearn.preprocessing import StandardScaler
X1 = StandardScaler().fit_transform(X)
```

```python
# Définition de l'architecture du modèle
import tensorflow as tf
#Initialisation du model
model = tf.keras.models.Sequential()
#1. Input layer
model.add(tf.keras.layers.Dense(7, input_dim=8 , activation='relu' )) 
# 2. Couche cachée
model.add(tf.keras.layers.Dense(5, activation='sigmoid'))
model.add(tf.keras.layers.Dense(4, activation='relu'))

model.add(tf.keras.layers.Dense(2, activation='relu'))

#3. Output layer
model.add(tf.keras.layers.Dense(2, activation='sigmoid'))
```

```python
# Division de l'ensemble de données en ensembles d'entraînement et de test
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X1, Y, test_size=0.2, random_state=42)
```

```python
# Compilation du modèle avec RMSprop comme optimiseur
model.compile(optimizer='RMSprop', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Entraînement du modèle
model.fit(X_train, Y_train, epochs=500, batch_size=32)
```

```python
# Évaluation du modèle
loss_rmsprop, accuracy_rmsprop = model.evaluate(X_test, Y_test)
print("Loss_rmsprop:", loss_rmsprop)
print("Accuracy_rmsprop:", accuracy_rmsprop)
```

```python
#Prédire le modele
Y_pred=model.predict(X_test)

from sklearn.metrics import accuracy_score
acc = accuracy_score(Y_test, Y_pred.round())
print('accuracy est', acc)
```

**Comparaison avec les autres optimisateurs** 


Lorsqu'il s'agit de comparer RMSprop avec d'autres optimisateurs en deep learning, il est important de prendre en compte différents aspects tels que la convergence, la stabilité, la vitesse et l'adaptabilité. Voici une comparaison de RMSprop avec deux optimisateurs populaires : l'optimiseur Stochastic Gradient Descent (SGD) et l'optimiseur Adam.

* Convergence : RMSprop a tendance à converger plus rapidement que SGD, en particulier dans les régions où les gradients sont bruyants ou les variations importantes. Cependant, Adam peut également converger rapidement en utilisant des estimations adaptatives des moments d'ordre supérieur.

* Stabilité : RMSprop est plus stable que SGD car il normalise les gradients et atténue les variations importantes. SGD peut être plus sujet aux oscillations et aux problèmes de convergence, en particulier avec des taux d'apprentissage élevés. Adam offre également une certaine stabilité en adaptant les taux d'apprentissage, mais il peut être légèrement moins stable que RMSprop.

* Vitesse : RMSprop a tendance à être plus rapide que SGD en raison de son adaptation des taux d'apprentissage. Cela lui permet de converger plus rapidement vers le minimum global. Adam peut également être rapide grâce à l'utilisation d'estimations adaptatives des moments.

* Adaptabilité : RMSprop est moins sensible aux choix des hyperparamètres que SGD, ce qui simplifie le processus d'optimisation. Adam a une meilleure adaptabilité grâce à l'adaptation des taux d'apprentissage et des moments d'ordre supérieur, mais il peut être plus sensible aux hyperparamètres que RMSprop.



```python
# Compilation du modèle avec SGD comme optimiseur
model.compile(optimizer = 'SGD', loss = 'binary_crossentropy', metrics = ['accuracy'] )
# Entraînement du modèle
model.fit(X_train, Y_train, epochs=500, batch_size=32)
```

```python
# Évaluation du modèle
loss_sgd, accuracy_sgd = model.evaluate(X_test, Y_test)
print("Loss_sgd:", loss_sgd)
print("Accuracy_sgd:", accuracy_sgd)
```

```python
# Compilation du modèle avec Adam comme optimiseureur
model.compile(optimizer = 'Adam', loss = 'binary_crossentropy', metrics = ['accuracy'] )
# Entraînement du modèle
model.fit(X_train, Y_train, epochs=500, batch_size=32)
```

```python
# Évaluation du modèle
loss_adam, accuracy_adam = model.evaluate(X_test, Y_test)
print("Loss_adam:", loss_adam)
print("Accuracy_adam:", accuracy_adam)
```

```python
import matplotlib.pyplot as plt

# Précisions obtenues avec RMSprop, SGD et Adam
accuracies = [accuracy_rmsprop, accuracy_sgd, accuracy_adam]
optimizers = ['RMSprop', 'SGD', 'Adam']
colors = ['blue', 'green', 'yellow']

# Tracé du graphique avec des couleurs différentes pour chaque optimiseur
plt.bar(optimizers, accuracies, color=colors)
plt.xlabel('Optimiseur')
plt.ylabel('Précision')
plt.title('Comparaison des performances des optimiseurs')
plt.ylim([0, 1])
plt.show()
```

### Utilisation du mini Batch Gradient Descent 1

## Avec le MLP

```python
from sklearn.datasets import fetch_openml

boston = fetch_openml(name='boston')
data = boston['data']
target = boston['target']
features = boston['feature_names']
```

```python
data.head()
```

```python
from sklearn.metrics.pairwise import kernel_metrics
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras import regularizers
def model_five_layers(dim_data):
  model=Sequential()
  model.add(Dense(100,input_dim=dim_data,activation='sigmoid',kernel_regularizer=regularizers.l2(0.01)))
  model.add(Dense(150,activation='relu'))
  model.add(Dropout(0.4))
  model.add(Dense(100,activation='sigmoid',kernel_regularizer=regularizers.l2(0.01)))
  model.add(Dropout(0.3))
  model.add(Dense(30,activation='sigmoid'))
  model.add(Dense(100,activation='sigmoid'))
  model.add(Dense(1,activation='linear'))
  model.compile(optimizer='SGD',loss='mean_squared_error')
  return model

```

```python
#preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_cr=StandardScaler().fit_transform(data)
X_train,X_test,Y_train,Y_test = train_test_split(X_cr,target,test_size = 0.3, random_state=0)

```

```python
#Apprentissage sans notre optimiseur
from keras.callbacks import EarlyStopping
callbacks=[EarlyStopping(monitor="val_loss",patience=50)]
model=model_five_layers(X_train.shape[1])
model.fit(X_train,Y_train,validation_data=(X_test,Y_test),epochs=500,callbacks=callbacks)
```

```python
#metrics de la premiere
from sklearn.metrics import mean_squared_error
mean_squared_error(model.predict(X_test),Y_test)
```

```python
#Apprentissage avec mini Batch gradient descent
# Entraînement du modèle avec mini-batch gradient descent

from keras.callbacks import EarlyStopping
callbacks=[EarlyStopping(monitor="val_loss",patience=50)]
model=model_five_layers(X_train.shape[1])
model.fit(X_train, Y_train, batch_size=5, epochs=200, validation_data=(X_test, Y_test))
```

```python
#metrics de la deuxiéme
from sklearn.metrics import mean_squared_error
mean_squared_error(model.predict(X_test),Y_test)
```
